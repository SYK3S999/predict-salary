<H2 style="background-color:powderblue;padding:20px;"> 1-Preprocessing </H2>

<H2 style="background-color:powderblue;padding:20px">we keep only some columns that we need {"Country","EdLevel","YearsCodePro","Employment","ConvertedCompYearly"} because we don't want the user to get bothered with the too much informations that he has to fill out</H2>

```python
df = df[["Country","EdLevel","YearsCodePro","Employment","ConvertedCompYearly"]]
df = df.rename({"ConvertedCompYearly":"salary"},axis=1)
df.head()
```

<H2 style="background-color:powderblue;padding:20px">Deletting the rows where the salary = null</H2>

```python
df=df[df['salary'].notnull()]
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
df.isnull().sum()
```

<H2 style="background-color:powderblue;padding:20px">we want to keep the rows where the employer is a full-time employer</H2>

```python
df = df[df['Employment'] == "Employed, full-time"]
df = df.drop('Employment',axis=1)
df.info()
```

```python
df["Country"].value_counts()
```

<H2 style="background-color:powderblue;padding:20px">now we make a function to delete the countries where the number of workers is less then <b style='color:red;'> 415</b></H2>

```python
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = 'other'
    return categorical_map
```

```python
country_map = shorten_categories(df.Country.value_counts(),415)
df['Country'] = df['Country'].map(country_map)
df.Country.value_counts()
```

<H2 style="background-color:powderblue;padding:20px">Now, we keep only the rows where the salary is <b style ='color:red;'> >= 10k </b>and  <b style ='color:red;'> <=250k</b>  </H2>

<H2 style="background-color:powderblue;padding:20px">noticing that we have string values in <b>"YearsCodePro"</b>, we should change it to float</H2>

```python
df['YearsCodePro'].unique()
```

```python
def clean_exeperience(x):
    if x == 'Less than 1 year':
        return 0.5
    if x == 'More than 50 years':
        return 50
    return float(x)
df["YearsCodePro"] = df['YearsCodePro'].apply(clean_exeperience)
df['YearsCodePro'].unique()
```

```python
df['EdLevel'].unique()
```

```python
def clean_education(x):
    if 'Bachelor’s degree' in x:
        return 'Bachelor’s degree'
    if 'Master’s degree' in x:
        return 'Master’s degree'
    if 'Professional degree' in x:
        return 'post grad'
    return 'less than a Bachelors'
df["EdLevel"] = df['EdLevel'].apply(clean_education)
df['EdLevel'].unique()
```

<H2 style="background-color:powderblue;padding:20px">we have some features of <b>String type</b>, but our model doesn't understand it , so it's time for encoding them </H2>

```python
from sklearn.preprocessing import LabelEncoder
le_education = LabelEncoder()
df['EdLevel'] = le_education.fit_transform(df['EdLevel'])
df['EdLevel'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le_country = LabelEncoder()
df['Country'] = le_country.fit_transform(df['Country'])
df['Country'].unique()
```
